```python
#Se importa la funcionalidad necesaria
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Se leen las tablas
publico=pd.read_csv('df_2017_publico_articulo.csv',sep=",")
publico_prestigio=pd.read_csv('df_2017_publico_prestigio.csv',sep=",")
cant_prestigio=pd.read_csv('df_2017_cant_prestigio.csv',sep=",")
personas_2017=pd.read_csv('personas_2017_sin_estandarizar.csv',sep=",")
```

```python
personas_2017_original=pd.read_csv('personas_2017.csv',sep=";")
```

```python
personas_2012_2016=pd.read_csv('personas_2012_2016.csv',sep=",")
```

```python
personas_2012_2016.head()
```

```python
personas_2017_original=personas_2017_original[['persona_id','disciplina_experticia_id']]
```

```python
personas_2017_original.head()
```

```python
disciplina=pd.read_csv('ref_disciplina.csv',sep=";")
disciplina.head()
```

```python
#Nos quedamos con Gran Área expertis
disciplina=disciplina[['disciplina_id','gran_area_descripcion']]
disciplina.head()
```

```python
#Hacemos merge con disciplina_experticia_id, renombramos columnas y eliminamos las que no se necesitan
personas_2017_original=pd.merge(personas_2017_original, disciplina, how='left',left_on='disciplina_experticia_id',right_on='disciplina_id')
personas_2017_original.rename(columns={"gran_area_descripcion": "area_experticia"},inplace=True)
personas_2017_original.drop(columns=['disciplina_experticia_id','disciplina_id'],inplace=True)
```
